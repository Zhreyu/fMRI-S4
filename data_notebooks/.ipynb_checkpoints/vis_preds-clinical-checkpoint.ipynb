```python
def calculate_score(data):
    
    X = data['output'].transpose(0,2,1)
    Y = data['target'].transpose(0,2,1)
    Z = data['y']

    
    mses = []
    corrs = []
    diff = []
    maes = [] 
    
    for (x,y) in zip(X, Y):
        
        x = torch.tensor(x)
        y = torch.tensor(y)
        corr = np.abs(pearsonr(x,y).numpy())#.mean()
        corr = corr.mean()
        mse = torch.mean(torch.square(torch.subtract(x,y)),axis=1).numpy().mean()
        mae = torch.mean(torch.abs(torch.subtract(x,y)),axis=1).numpy().mean()

        corrs.append(corr)
        mses.append(mse)
        maes.append(mae)
        diff.append(-corr)
    return mses, maes, corrs, diff,  Z


    
    
 
    
```

```python
mses, maes, corrs, diff, labels = calculate_score(data)
auc = roc_auc_score(labels, diff)
auc
```

```python
def calculate_score_roi(data,roi):
    
    X = data['output'].transpose(0,2,1)[:,roi,:]
    Y = data['target'].transpose(0,2,1)[:,roi,:]
    Z = data['y']

    
    mses = []
    corrs = []
    diff = []
    maes = [] 
    
    for (x,y) in zip(X, Y):
        
        x = torch.tensor(x)
        y = torch.tensor(y)
        corr = np.abs(pearsonr(x,y).numpy()).mean()
        mse = torch.mean(torch.square(torch.subtract(x,y))).numpy().mean()
        mae = torch.mean(torch.abs(torch.subtract(x,y))).numpy().mean()

        corrs.append(corr)
        mses.append(mse)
        maes.append(mae)
        diff.append(corr-mse)
    return mses, maes, corrs, diff,  Z


    
    
 
    
```

```python
from nilearn import datasets

dataset = datasets.fetch_atlas_harvard_oxford('sub-maxprob-thr25-2mm', symmetric_split= True)
rois_ho = dataset.labels[1:]
rois_important = []
for i,roi in enumerate(rois_ho):
    mses, maes, corrs, diff, labels = calculate_score_roi(data,i)
    auc = roc_auc_score(labels, mses)
    if auc> 0.56:
        rois_important.append(i)
    print('{}  auc = {:.3f}'.format(roi, auc))
```

```python
from sklearn.manifold import TSNE
def plot_tnse(anomaly_output, normal_output):
    
    Y_a = anomaly_output['target'].transpose(0,2,1)
    X_a = anomaly_output['output'].transpose(0,2,1)
    Z_a = np.ones(Y_a.shape[0])
    
    Y_n = normal_output['target'].transpose(0,2,1)
    X_n = normal_output['output'].transpose(0,2,1)
    Z_n = np.zeros(Y_n.shape[0])
    
    Y = np.concatenate([Y_a, Y_n],axis=0)
    X = np.concatenate([X_a, X_n],axis=0)
    Z = np.concatenate([Z_a, Z_n],axis=0)
    
    mses = []
    maes = [] 
    
    for (x,y) in zip(X, Y):
        
        x = torch.tensor(x)
        y = torch.tensor(y)
        mse = torch.mean(torch.square(torch.subtract(x,y)),axis=1).numpy()
        mae = torch.mean(torch.abs(torch.subtract(x,y)),axis=1).numpy()
        
        mses.append(mse)
        maes.append(mae)
        
    mses = np.array(mses)
    maes = np.array(maes)
    
    tsne = TSNE(n_components=2, learning_rate='auto', init='random', verbose=False)
    tsne_results = tsne.fit_transform(maes)
    
    return tsne_results, Z


    
    
 
    
```

```python
import seaborn as sns
tsne_resutls, labels = plot_tnse(anomaly_ouptut, normal_ouptut)
df = pd.DataFrame(tsne_resutls, columns=['x','y'])
df['labels'] = labels
df['labels'].replace({1:'MDD', 0:'Healthy'},inplace=True)
fig = plt.figure(figsize=(16,10))
snsplot = sns.scatterplot(
    x="x", y="y",
    hue="labels",
    data=df,
    legend="full",s=100
).set(title='Tsne plot of ROI MSE erros in predicting 150 timepoints')

plt.savefig('tsne-forecast_150.png')
```

```python
def plot_subject(ax, subject, roi=-1, name='ground_truth'):
    if roi == -1:
        for region in subject:
            ax.plot(region)
    else:
        for region in subject[roi:roi+1]:
            ax.plot(region, label=name)
```
